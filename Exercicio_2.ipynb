### a) 
Considere os polígonos dados pelas suas componentes convexas:
$$

$$
e
$$
P_{2} = \begin{bmatrix} \begin{bmatrix} (8,2), (9,2), (9,3) \end{bmatrix}, \begin{bmatrix} (9,1), (10,2), (9,3) \end{bmatrix} \end{bmatrix}.
$$

Qual a menor faixa retangular com a qual seja possível dispor os objetos sem que haja sobreposição?

```julia
# Plota o problema
poligonos = [
    [[1.0 4.0; 4.0 1.0; 4.0 3.0], [4.0 1.0; 9.0 7.0; 4.0 3.0]], 
    [[8.0 2.0; 9.0 2.0; 9.0 3.0], [9.0 1.0; 10.0 2.0; 9.0 3.0]]
]
W = 10
L = 10
circulos_r = []
circulos_c = []

#plota_problema(circulos_c, circulos_r, poligonos, 10, 10)
```

```julia

n = size(poligonos[1][1])[1]
P_refe = zeros(n, 2) 
 for i=1:n
    P_refe[i, :] =(poligonos[1][1])[i, :] - (poligonos[1][1])[1, :]
 end

reta = P_refe[1:2, :]

n_2 = size(poligonos[1][2])[1]
P_refe_2 = zeros(n, 2) 
 for i=1:n_2
    P_refe_2[i, :] =(poligonos[1][2])[i, :] - (poligonos[1][2])[1, :]
 end

reta_2 = P_refe_2[1:2, :]

n_3 = size(poligonos[2][1])[1]
P_refe_3= zeros(n_3, 2) 
 for i=1:n_3
    P_refe_3[i, :] =(poligonos[2][1])[i, :] - (poligonos[2][1])[1, :]
 end

reta_3 = P_refe_3[1:2, :]


n_4 = size(poligonos[2][2])[1]
P_refe_4= zeros(n_3, 2) 
 for i=1:n_4
    P_refe_4[i, :] =(poligonos[2][2])[i, :] - (poligonos[2][2])[1, :]
 end

reta_4 = P_refe_4[1:2, :]

#Modelo Criado
model = Model(Ipopt.Optimizer)
@variable(model, xP[1:2])
@variable(model, yP[1:2])
@variable(model, theta_P[1:2])
@variable(model, xL)
@variable(model, yL)
@variable(model, alpha_L)
@NLexpression(model, NLe1[i=1:n], 
    P_refe[i, 1] * cos(theta_P[1]) - P_refe[i, 2] * sin(theta_P[1]) + xP[1])
@NLexpression(model, NLe2[i=1:n], 
    P_refe[i, 1] * sin(theta_P[1]) + P_refe[i, 2] * cos(theta_P[1]) + yP[1])
@NLconstraint(model, [i=1:n], 0.0 <= NLe1[i] <= W )
@NLconstraint(model, [i=1:n], 0.0 <= NLe2[i] <= L )

@NLexpression(model, NLe3[j=1:n_2], 
    P_refe_2[j, 1] * cos(theta_P[1]) - P_refe_2[j, 2] * sin(theta_P[1]) + xP[1])
@NLexpression(model, NLe4[j=1:n_2], 
    P_refe_2[j, 1] * sin(theta_P[1]) + P_refe_2[j, 2] * cos(theta_P[1]) + yP[1])
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe3[j] <= W )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe4[j] <= L )

@NLexpression(model, NLe5[j=1:n_2], 
    P_refe_3[j, 1] * cos(theta_P[2]) - P_refe_3[j, 2] * sin(theta_P[2]) + xP[2])
@NLexpression(model, NLe6[j=1:n_2], 
    P_refe_3[j, 1] * sin(theta_P[2]) + P_refe_3[j, 2] * cos(theta_P[2]) + yP[2])
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe5[j] <= W )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe6[j] <= L )

@NLexpression(model, NLe7[j=1:n_2], 
    P_refe_4[j, 1] * cos(theta_P[2]) - P_refe_4[j, 2] * sin(theta_P[2]) + xP[2])
@NLexpression(model, NLe8[j=1:n_2], 
    P_refe_4[j, 1] * sin(theta_P[2]) + P_refe_4[j, 2] * cos(theta_P[2]) + yP[2])
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe5[j] <= W )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe6[j] <= L )







@NLexpression(model, cij, 
    ((reta[2, 1] - reta[1, 1]) * sin(alpha_L) + (reta[2, 2] - reta[1, 2]) * cos(alpha_L)) /
    ((reta[2, 1] - reta[1, 1]) * cos(alpha_L) + (reta[1, 2] - reta[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij, yL - cij * xL)


@NLexpression(model, cij_2, 
    ((reta_2[2, 1] - reta_2[1, 1]) * sin(alpha_L) + (reta_2[2, 2] - reta_2[1, 2]) * cos(alpha_L)) /
    ((reta_2[2, 1] - reta_2[1, 1]) * cos(alpha_L) + (reta_2[1, 2] - reta_2[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij_2, yL - cij_2 * xL)


@NLexpression(model, cij_3, 
    ((reta_3[2, 1] - reta_3[1, 1]) * sin(alpha_L) + (reta_3[2, 2] - reta_3[1, 2]) * cos(alpha_L)) /
    ((reta_3[2, 1] - reta_3[1, 1]) * cos(alpha_L) + (reta_3[1, 2] - reta_3[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij_3, yL - cij_3 * xL)


@NLexpression(model, cij_4, 
    ((reta_3[2, 1] - reta_4[1, 1]) * sin(alpha_L) + (reta_4[2, 2] - reta_4[1, 2]) * cos(alpha_L)) /
    ((reta_3[2, 1] - reta_4[1, 1]) * cos(alpha_L) + (reta_4[1, 2] - reta_4[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij_4, yL - cij_4 * xL)

@NLconstraint(model, [i=1:n], NLe6[i] - cij * NLe1[i] - dij <= 0.0)
@NLconstraint(model, [j=1:n_2], NLe7[j] - cij * NLe3[j] - dij >= 0.0)


@NLconstraint(model, [i=1:n], NLe6[i] - cij_2 * NLe3[i] - dij_2 <= 0.0)
@NLconstraint(model, [j=1:n_2], NLe7[j] - cij_2 * NLe4[j] - dij_2>= 0.0)



@NLconstraint(model, [i=1:n], NLe3[i] - cij_3 * NLe1[i] - dij_3 <= 0.0)
@NLconstraint(model, [j=1:n_2], NLe4[j] - cij_2 * NLe2[j] - dij_2>= 0.0)

@objective(model, Min, 0)

# Chama o otimizador Ipopt:
status = optimize!(model)
x= [value.(xP)[1], value.(yP)[1], value.(theta_P)[1],value.(xP)[2], value.(yP)[2], value.(theta_P)[2]]

using Plots

#plota_solucao(lista_circulos, lista_poligonos,L, W,x)

```

### b) 
Considere os polígonos dados pelas suas componentes convexas:
$$
P_{1} = \begin{bmatrix} \begin{bmatrix}(1,4), (4, 1), (4, 3)\end{bmatrix}, \begin{bmatrix} (4,1), (9,7), (4,3) \end{bmatrix} \end{bmatrix},
$$
e
$$
P_{2} = \begin{bmatrix} \begin{bmatrix} (8,2), (9,2), (9,3) \end{bmatrix}, \begin{bmatrix} (9,1), (10,2), (9,3) \end{bmatrix} \end{bmatrix}.
$$
e o círculo de raio $2$ e centro $(13,3)$.

Qual a menor faixa retangular com a qual seja possível dispor os objetos sem que haja sobreposição?

```julia
circulos_r = [2]
circulos_c = [13,3]
n = size(poligonos[1][1])[1]
P_refe = zeros(n, 2) 
 for i=1:n
    P_refe[i, :] =(poligonos[1][1])[i, :] - (poligonos[1][1])[1, :]
 end

reta = P_refe[1:2, :]

n_2 = size(poligonos[1][2])[1]
P_refe_2 = zeros(n, 2) 
 for i=1:n_2
    P_refe_2[i, :] =(poligonos[1][2])[i, :] - (poligonos[1][2])[1, :]
 end

reta_2 = P_refe_2[1:2, :]

n_3 = size(poligonos[2][1])[1]
P_refe_3= zeros(n_3, 2) 
 for i=1:n_3
    P_refe_3[i, :] =(poligonos[2][1])[i, :] - (poligonos[2][1])[1, :]
 end

reta_3 = P_refe_3[1:2, :]


n_4 = size(poligonos[2][2])[1]
P_refe_4= zeros(n_3, 2) 
 for i=1:n_4
    P_refe_4[i, :] =(poligonos[2][2])[i, :] - (poligonos[2][2])[1, :]
 end

reta_4 = P_refe_4[1:2, :]

#Modelo Criado
#Variavéis
model = Model(Ipopt.Optimizer)
@variable(model, L>0)
@variable(model,W>0)
@variable(model, xP[1:2])
@variable(model, yP[1:2])
@variable(model, theta_P[1:2])
@variable(model,xC[1:2])
@variable(model,yC[1:2])
@variable(model, xL)
@variable(model, yL)
@variable(model, alpha_L)




#expressões não lineares
@NLexpression(model, NLe1[i=1:n], 
    P_refe[i, 1] * cos(theta_P[1]) - P_refe[i, 2] * sin(theta_P[1]) + xP[1] )
@NLexpression(model, NLe2[i=1:n], 
    P_refe[i, 1] * sin(theta_P[1]) + P_refe[i, 2] * cos(theta_P[1]) + yP[1] )

@NLexpression(model, NLe3[j=1:n_2], 
    P_refe_2[j, 1] * cos(theta_P[1]) - P_refe_2[j, 2] * sin(theta_P[1]) + xP[1]- W)
@NLexpression(model, NLe4[j=1:n_2], 
    P_refe_2[j, 1] * sin(theta_P[1]) + P_refe_2[j, 2] * cos(theta_P[1]) + yP[1]-L)
@NLexpression(model, NLe5[j=1:n_2], 
    P_refe_3[j, 1] * cos(theta_P[2]) - P_refe_3[j, 2] * sin(theta_P[2]) + xP[2]-W)
@NLexpression(model, NLe6[j=1:n_2], 
    P_refe_3[j, 1] * sin(theta_P[2]) + P_refe_3[j, 2] * cos(theta_P[2]) + yP[2]-L)
@NLexpression(model, NLe7[j=1:n_2], 
    P_refe_4[j, 1] * cos(theta_P[2]) - P_refe_4[j, 2] * sin(theta_P[2]) + xP[2]-W)
@NLexpression(model, NLe8[j=1:n_2], 
    P_refe_4[j, 1] * sin(theta_P[2]) + P_refe_4[j, 2] * cos(theta_P[2]) + yP[2]-L)
@NLexpression(model, cij, 
    ((reta[2, 1] - reta[1, 1]) * sin(alpha_L) + (reta[2, 2] - reta[1, 2]) * cos(alpha_L)) /
    ((reta[2, 1] - reta[1, 1]) * cos(alpha_L) + (reta[1, 2] - reta[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij, yL - cij * xL)
@NLexpression(model, cij_2, 
    ((reta_2[2, 1] - reta_2[1, 1]) * sin(alpha_L) + (reta_2[2, 2] - reta_2[1, 2]) * cos(alpha_L)) /
    ((reta_2[2, 1] - reta_2[1, 1]) * cos(alpha_L) + (reta_2[1, 2] - reta_2[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij_2, yL - cij_2 * xL)
@NLexpression(model, cij_3, 
    ((reta_3[2, 1] - reta_3[1, 1]) * sin(alpha_L) + (reta_3[2, 2] - reta_3[1, 2]) * cos(alpha_L)) /
    ((reta_3[2, 1] - reta_3[1, 1]) * cos(alpha_L) + (reta_3[1, 2] - reta_3[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij_3, yL - cij_3 * xL)
@NLexpression(model, cij_4, 
    ((reta_3[2, 1] - reta_4[1, 1]) * sin(alpha_L) + (reta_4[2, 2] - reta_4[1, 2]) * cos(alpha_L)) /
    ((reta_3[2, 1] - reta_4[1, 1]) * cos(alpha_L) + (reta_4[1, 2] - reta_4[2, 2]) * sin(alpha_L)) )
@NLexpression(model, dij_4, yL - cij_4 * xL)
@NLexpression(model, delta_ij[1:j], abs(cij[j] * xC[1] - yC[1] + dij[j]) / sqrt(cij[j] ^ 2.0 + 1.0))
@NLexpression(model, delta_ij_2[1:j], abs(cij[j] * xC[2] - yC[2] + dij[j]) / sqrt(cij[j] ^ 2.0 + 1.0))


#constraint de contenção
@constraint(model, [i=1:2], raios[i] - xC[i] <= 0.0)
@constraint(model, [i=1:2], raios[i] + xC[i] <= W)
@constraint(model, [i=1:2], raios[i] - yC[i] <= 0.0)
@constraint(model, [i=1:2], raios[i] + yC[i] <= L)
@NLconstraint(model, [i=1:n], 0.0 <=NLe1[i]-W )
@NLconstraint(model, [i=1:n], 0.0 <= NLe2[i]-L )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe3[j] -W  )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe4[j]-L  )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe5[j]-W  )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe6[j] -L)
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe7[j]-W  )
@NLconstraint(model, [j=1:n_2], 0.0 <= NLe8[j] -L )
@NLconstraint(model,[1:j],delta_ij>=r[1])
@NLconstraint(model,[1:j],delta_ij_2>=r[2])









#constraint de não sobreposição
@NLconstraint(model, [i=1:n_3], NLe6[i] - cij * NLe1[i] - dij <= 0.0)
@NLconstraint(model, [j=1:n_4], NLe7[j] - cij * NLe3[j] - dij >= 0.0)
@NLconstraint(model, [i=1:n_2], NLe6[i] - cij_2 * NLe3[i] - dij_2 <= 0.0)
@NLconstraint(model, [j=1:n_3], NLe7[j] - cij_2 * NLe4[j] - dij_2>= 0.0)
@NLconstraint(model, [i=1:n], NLe3[i] - cij_3 * NLe1[i] - dij_3 <= 0.0)
@NLconstraint(model, [j=1:n_2], NLe4[j] - cij_2 * NLe2[j] - dij_2>= 0.0)
@NLconstraint(model, (xC[1] - xC[2])^2.0 + (yC[1] - yC[2])^2.0 >= (raios[1] + raios[2])^2.0)




@NLobjective(model, Min,L*W)

# Chama o otimizador Ipopt:
status = optimize!(model)
```
